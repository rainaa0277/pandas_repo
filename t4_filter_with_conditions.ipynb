```python
import pandas as pd
```

```python
person = {"first": ["Rishab", "Rajas", "Sam"],
          "last": ["Raina", "Raina", "Sulek"],
          "email": ["rainaa0277@gmail.com", "rajasraut@yahoo.com", "samsulek01@gmail.com"]}
```

```python
person_df = pd.DataFrame(person)
person_df
```

```python
(person_df['last'] == 'Raina')
```

```python
#filter people who have 'Raina' as last name
filt = (person_df['last'] == 'Raina')
```

```python
person_df[filt]
```

```python
person_df.loc[filt,"email"]   #loc works for boolean
```

## Operators
#### & - AND
#### | - OR

```python
#filter rows where last = 'Raina' and first = 'Rishab'
filt = (person_df['last'] == 'Raina') & (person_df['first'] == 'Rishab')
```

```python
person_df.loc[filt]
```

```python
#filter rows where last = 'Raina' OR first = 'Rishab'
filt = (person_df['last'] == 'Raina') | (person_df['first'] == 'Rishab')
```

```python
person_df.loc[filt]
```

```python
#filter rows where neither last = 'Raina' nor first = 'Rishab'
filt = (person_df['last'] == 'Raina') | (person_df['first'] == 'Rishab')
```

```python
person_df.loc[~filt] #tilda is used for opposite condition
```
